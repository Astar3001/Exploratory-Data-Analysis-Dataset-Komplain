Ambil data Komplain tahun 2020-2023

```python
df['Tanggal'] = pd.to_datetime(df['Tanggal'], format='%d/%m/%Y')
```

```python
df['tahun'] = df['Tanggal'].dt.year
```

```python
filtered_df = df[df['tahun'].isin([2020, 2021, 2022, 2023])]
display(filtered_df.head())
```

Ambil data Kependudukan tahun 2020-2023

```python
fil_data = df[df['tahun'].isin([2020, 2021, 2022, 2023])]
```

```python
fil_data.head()
```

Value pada kolom indikator

```python

complaint_counts_by_year = filtered_df.pivot_table(
    index="tahun",
    columns="Alasan Komplain",
    aggfunc='size',
    fill_value=0
)

display(complaint_counts_by_year)
```

```python
total = fil_data.groupby('tahun').size().reset_index(name='jumlah_komplain')
display(total)
```

Menyatukan kedua DataFrame berdasarkan tahun

```python
# Merging the two DataFrames based on the 'tahun' column
merged_df = pd.merge(complaint_counts_by_year, total, on='tahun')
display(merged_df)
```

Data Exploration & Data Visualisation

```python
# Mengganti visualisasi kepadatan penduduk dengan visualisasi total keluhan per tahun
# Karena kolom asli tidak ditemukan dalam dataset keluhan

# Mengambil data total keluhan per tahun dari DataFrame 'total'
# DataFrame 'total' sudah dihitung di sel sebelumnya

# Membuat bar chart total keluhan per tahun
plt.figure(figsize=(10, 6)) # Mengatur ukuran plot
bars = plt.bar(total['tahun'], total['jumlah_komplain']) # Membuat bar chart

plt.xlabel('Tahun') # Label sumbu x
plt.ylabel('Jumlah Komplain') # Label sumbu y
plt.title('Total Jumlah Komplain per Tahun (2020-2023)') # Judul plot
plt.xticks(total['tahun']) # Menampilkan semua tahun di sumbu x

# Loop untuk menambahkan keterangan nilai di setiap batang grafik
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, yval, va='bottom', ha='center')

plt.show() # Menampilkan plot
```

Melihat tren Indeks Kualitas

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Menggunakan DataFrame 'total' yang berisi total keluhan per tahun
# Original code was for IKLH data which is not available

plt.figure(figsize=(10, 6))
sns.lineplot(data=total, x="tahun", y="jumlah_komplain")
plt.title('Tren Total Jumlah Komplain dari Tahun ke Tahun (2020-2023)')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Komplain')
plt.xticks(total['tahun']) # Memastikan semua tahun ditampilkan di sumbu x

# Menambahkan keterangan nilai di setiap titik grafik
for i in range(len(total)):
    plt.annotate(total.iloc[i]['jumlah_komplain'],
                 (total.iloc[i]['tahun'], total.iloc[i]['jumlah_komplain']),
                 textcoords="offset points",
                 xytext=(0,10),
                 ha='center')

plt.show()
```

```python
# Mengganti pemfilteran berdasarkan 'indikator' dengan pemfilteran berdasarkan 'tahun'
# Karena kolom 'indikator' tidak ditemukan dalam dataset keluhan.
# Hasil pemfilteran ini akan serupa dengan filtered_df.
df_ind = filtered_df[filtered_df['tahun'].isin([2020, 2021, 2022, 2023])]

# Menampilkan beberapa baris pertama dari DataFrame yang difilter
display(df_ind.head())
```

```python
plt.figure(figsize=(14, 7))

# Memilih beberapa alasan komplain teratas untuk divisualisasikan
# Anda bisa menyesuaikan daftar ini sesuai kebutuhan
top_complaints = complaint_counts_by_year.sum().sort_values(ascending=False).head(5).index.tolist()

# Mengubah format merged_df agar sesuai untuk seaborn lineplot (long format)
merged_df_long = merged_df.melt(
    id_vars='tahun',
    value_vars=top_complaints, # Memilih kolom alasan komplain teratas
    var_name='Alasan Komplain',
    value_name='Jumlah Komplain'
)

# Membuat line plot tren setiap alasan komplain teratas
ax = sns.lineplot(data=merged_df_long, x="tahun", y="Jumlah Komplain", hue="Alasan Komplain", marker='o')

plt.title('Tren Jumlah Komplain untuk Beberapa Alasan Teratas per Tahun (2020-2023)')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Komplain')
plt.xticks(merged_df_long['tahun'].unique()) # Menampilkan semua tahun di sumbu x
plt.grid(True, linestyle='--', alpha=0.6) # Menambahkan grid

# Menambahkan keterangan nilai di setiap titik grafik (opsional, bisa membuat ramai)
# for i in range(len(merged_df_long)):
#     plt.annotate(merged_df_long['Jumlah Komplain'].iloc[i],
#                  (merged_df_long['tahun'].iloc[i], merged_df_long['Jumlah Komplain'].iloc[i]),
#                  textcoords="offset points",
#                  xytext=(0,10),
#                  ha='center')

plt.tight_layout() # Menyesuaikan layout
plt.show()
```

```python
# Membuat heatmap korelasi Pearson
plt.figure(figsize=(10, 8))
ax = sns.heatmap(merged_df.corr(), annot=True, fmt=".2f", cmap='coolwarm')

# Menambahkan judul
plt.title('Heatmap Korelasi Pearson Antara Fitur-fitur')

# Menghilangkan label sumbu x dan y
ax.set(xlabel="", ylabel="")

plt.show()
```

```python
# Menghitung korelasi Kendall antara fitur-fitur dalam merged_df
kendall_corr = merged_df.corr(method='kendall')

# Menampilkan heatmap korelasi Kendall
plt.figure(figsize=(10, 8))
ax = sns.heatmap(kendall_corr, annot=True, fmt=".2f", cmap='coolwarm')

# Menambahkan judul
plt.title('Heatmap Korelasi Kendall Antara Fitur-fitur')

# Menghilangkan label sumbu x dan y
ax.set(xlabel="", ylabel="")

plt.show()
```

```python
# Menghitung korelasi Spearman antara fitur-fitur dalam merged_df
spearman_corr = merged_df.corr(method='spearman')

# Menampilkan heatmap korelasi Spearman
plt.figure(figsize=(10, 8))
ax = sns.heatmap(spearman_corr, annot=True, fmt=".2f", cmap='coolwarm')

# Menambahkan judul
plt.title('Heatmap Korelasi Spearman Antara Fitur-fitur')

# Menghilangkan label sumbu x dan y
ax.set(xlabel="", ylabel="")

plt.show()
```

```python
# Compute correlation matrices
corr_pearson = merged_df.corr(method='pearson')
corr_kendall = merged_df.corr(method='kendall')
corr_spearman = merged_df.corr(method='spearman')

# Plot correlation matrices
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.heatmap(corr_pearson, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Pearson Correlation')

plt.subplot(1, 3, 2)
sns.heatmap(corr_kendall, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Kendall Correlation')

plt.subplot(1, 3, 3)
sns.heatmap(corr_spearman, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Spearman Correlation')

plt.tight_layout()
plt.show()
```
